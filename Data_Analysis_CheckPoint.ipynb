<h2>ניקוי המידע </h2>
<p>
ראינו שיש לנו תאים שפוגעים ומשנים את הניתוח שלנו על הנתונים. </br>
כמו תאים חסרים או ריקים, עמודות לא רלוונטיות </br>
לדוגמה במקומות מסוימים היו לנו נתונים שגויים (כמו גיל 0)

```python
#Droping the irrelevant columns   
columns_to_drop = [ 'Crm Cd 1', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Rpt Dist No', 'Part 1-2', 'Crm Cd','Premis cd','Cross Street','Weapon Used Cd','Status']
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
#Filtering out missing data 
df = df[~(df['Vict Sex'].apply(lambda x: pd.isnull(x) or str(x).strip() == '') |
        df['Vict Descent'].apply(lambda x: pd.isnull(x) or str(x).strip() == ''))]
df = df[df['Vict Age'] > 0]
```

<h2>עיבוד ראשוני של הדאטה</h2>


```python
'''
Decompositioning the column of Date Rptd to the different time parts (year, month,day)
df['Hour'] made by taking the only the hour from the time of occurrence 
'''
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

# Extract the year, month, and day into separate columns
df['Year'] = df['Date Rptd'].dt.year
df['Month'] = df['Date Rptd'].dt.month
df['Day'] = df['Date Rptd'].dt.day

```

```python
# Extract hour from 'TIME OCC' using loc, assuming 'TIME OCC' is in 4-digit format
df['Hour'] = None  # Initialize column
df.loc[df['TIME OCC'].notna(), 'Hour'] = df.loc[df['TIME OCC'].notna(), 'TIME OCC'] // 100
```

```python
# Create 'Vict Age Cat' column age categories with loc 
df['Vict Age Cat'] = None  # Initialize column
df.loc[df['Vict Age'] <= 4, 'Vict Age Cat'] = 'Baby'
df.loc[(df['Vict Age'] > 4) & (df['Vict Age'] <= 12), 'Vict Age Cat'] = 'Child'
df.loc[(df['Vict Age'] > 12) & (df['Vict Age'] <= 19), 'Vict Age Cat'] = 'Teen'
df.loc[(df['Vict Age'] > 19) & (df['Vict Age'] <= 39), 'Vict Age Cat'] = 'Adult'
df.loc[(df['Vict Age'] > 39) & (df['Vict Age'] <= 59), 'Vict Age Cat'] = 'Middle'
df.loc[df['Vict Age'] > 59, 'Vict Age Cat'] = 'Senior'
```

<h3>הוספת עמודת עונות</h3>


```python
# Assign seasons using loc and month values
df['Season'] = None  # Initialize column
df.loc[df['Month'].isin([12, 1, 2]), 'Season'] = 'Winter'
df.loc[df['Month'].isin([3, 4, 5]), 'Season'] = 'Spring'
df.loc[df['Month'].isin([6, 7, 8]), 'Season'] = 'Summer'
df.loc[df['Month'].isin([9, 10, 11]), 'Season'] = 'Fall'

df
```

<h3>חלוקת היום לזמנים כללים </h3>

```python
df['Day_Part'] = 'Night'  # ברירת מחדל היא לילה
df.loc[(df['Hour'] >= 6) & (df['Hour'] < 12), 'Day_Part'] = 'Morning'   # בוקר
df.loc[(df['Hour'] >= 12) & (df['Hour'] < 18), 'Day_Part'] = 'Afternoon' # צהריים
df.loc[(df['Hour'] >= 18) & (df['Hour'] < 22), 'Day_Part'] = 'Evening'   # ערב
data = df
df[['Day_Part']]
```

```python
# יצירת טווחי גילאים
bins = [0, 17, 35, 50, 120]  # גבולות טווחי הגיל
labels = ['0-17', '18-35', '36-50', '51+']  # שמות הטווחים
df['Age Group'] = pd.cut(df['Vict Age'], bins=bins, labels=labels, right=False)

# ספירת הפשעים לפי קבוצות גיל וסוג הפשע
crime_age_analysis = df.groupby(['Age Group', 'Crime Category'], observed=False).size().unstack(fill_value=0)

# הצגת התוצאה
crime_age_analysis
```

<h2>לפי עונות שנה </h2>

```python
# Extract relevant columns for the new analysis
crime_data = data[['Season', 'Crime Category']]
# Exclude rows where 'Crime Category' is 'Other'
crime_data = crime_data[crime_data['Crime Category'] != 'Other']
# Group by Season and Crime Description and calculate counts
crime_counts = crime_data.groupby(['Season', 'Crime Category']).size().reset_index(name='Count')

# Sort within each season to get the top 5 crimes
top_crimes = (crime_counts.groupby('Season', group_keys=False).apply(lambda x: x.nlargest(5, 'Count')).reset_index(drop=True))
# Calculate the percentages
total_counts = top_crimes.groupby('Season')['Count'].transform('sum')
top_crimes['Percentage'] = (top_crimes['Count'] / total_counts) * 100

# Plotting bar graphs for top 5 crimes in each season with percentages
plt.figure(figsize=(14, 8))
barplot = sns.barplot(data=top_crimes, x='Season', y='Count', hue='Crime Category', palette='pastel')

# Adding percentages on top of the bars
for bar, percentage in zip(barplot.patches, top_crimes['Percentage']):
    bar_height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar_height + 2,  # Adjusting the position slightly above the bar
        f'{percentage:.1f}%',
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.title('Top 5 Crimes in Each Season (with Percentages)', fontsize=16)
plt.xlabel('Season', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Crime Description', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
